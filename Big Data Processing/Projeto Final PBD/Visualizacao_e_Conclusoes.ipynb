Este notebook descreve o processo de desenvolvimento e os resultados do projeto de Processamento de Big Data. Este projeto tem como objetivo, a utilização dos materiais e métodos especificados nas aulas práticas desta mesma cadeira.

## **1. Definição do Problema**


### 1.1. Compreensão do *Dataset*


Para o desenvolvimento deste projeto foi selecionada a base de dados *“Smart Meters in London”*, obtida a partir do portal *Kaggle*. Este disponibiliza informações detalhadas sobre o consumo de energia de diversas residências em Londres (diário, por hora ou de meia em meia hora), juntamente com os perfis *Acorn*, que categorizam as residências de acordo com fatores socioeconômicos. Também são disponibilizados *csv's* sobre os feriados bancários do Reino Unido e sobre condições meteorológicas.


A seleção deste dataset deveu-se à quantidade e diversidade do mesmo. A quantidade de *csv's* e a diversidade dos mesmos, concede-lhe muito potencial para a resolução dos mais diversos problemas.


Com o aumento contínuo da quantidade de dados disponíveis, a análise de grandes volumes de informação tornou-se essencial para a obtenção de *insights* valiosos e a tomada de decisões informadas. Neste contexto, a capacidade de processamento e análise de grandes conjuntos de dados sobre o consumo de energia, pode oferecer uma compreensão mais profunda dos hábitos dos consumidores, permitindo uma melhor gestão da demanda energética, e também a criação de estratégias personalizadas para diferentes grupos de consumidores.   

<br>

### 1.2. Construção do Problema 

Para a definição do problema é necessário que o mesmo esteja de acordo com o *dataset* escolhido. Para tal, primeiramente foi feita a compreensão do *dataset* (*“Smart Meters in London”*) para, a partir daí, construir um problema adequado. 

<br>

Definições de alguns conceitos:
- *block*: Os blocos apenas servem para dividir o *dataset* original em vários mais pequenos. Esta divisão é feita de forma aleatória. [1]
- *Acorn*: Abreviação de *“A Classification of Residential Neighbourhoods”*, ou seja, é um sistema de classificação de áreas residenciais que se baseia no poder de compra e no estatuto social das pessoas que aí residem. A *Acorn* é uma poderosa ferramenta de classificação dos consumidores que segmenta a população do Reino Unido por código postal. [2]

<br>

Compreensão dos ficheiros contidos na base de dados: 

- *informations_households.csv*: este ficheiro contém informação sobre as residências onde foram registadas as contagens energéticas com os *smart meters*.

- *Daily_dataset.zip*, *hourly_dataset.zip* e *halfhourly_dataset.zip*Estes  ficheiros contêm os registos das contagens energéticas. Apenas diferem na frequência dos registros, sendo que o primeiro contém registos diários, o segundo registos de hora em hora e o terceiro registos de meia em meia hora. Os três ficheiros estão separados por *blocks*, no entanto, existe também um ficheiro chamado *daily_dataset.csv*, que contém os registos energéticos diários, sem a divisão dos *blocks*.    

- *acorn_details.csv*: este ficheiro compreende os detalhes dos grupos acorn e o respetivo perfil das pessoas pertencentes a cada grupo. Os detalhes de determinado acorn (X) são representados numa coluna específica com o nome ACORN-X. Numa escala nacional, o índice é 100, se, para determinada coluna o valor for 150, isso significa que existem 1.5 vezes mais pessoas com esse atributo no grupo acorn do que na escala nacional.

- *weather_daily_darksky.csv* e *weather_hourly_darksky.csv*: estes ficheiros contêm a informação diária e de hora a hora, respetivamente, sobre as condições meteorológicas em Londres. 

<br>

O problema definido para este projeto é a identificação de perfis de consumo energético, para posteriormente os comparar com fatores socioeconómicos disponibilizados na base de dados. A resolução deste problema implica a utilização de técnicas de aprendizagem não supervisionada, ou, mas especificamente, técnicas de *clustering*.

<br>

A abordagem para alcançar este objetivo, seguiu as seguintes etapas: 


- **Pré-processamento dos Dados**: Nesta etapa  foi realizada a extração, transformação e carregamento (ETL) dos dados;


- **Análise Exploratória dos Dados**: Inicialmente foi feita a compreensão e análise da relevância das variáveis no contexto do problema. Seguidamente foi realizada uma análise individual e conjunta das variáveis, dando maior importância e profundidade à análise das variáveis identificadas como mais relevantes. Por fim foram observadas as correlações entre as variáveis numéricas.


- **Modelação**: Incluiu a preparação para a modelação, a construção de modelos e a criação de pipelines para os modelos construídos;


- **Treinamento e Avaliação dos Modelos**: Nesta etapa foram testados os modelos criados no conjunto de teste guardado anteriormente, e, através da visualização dos gráficos resultantes, foi feita a interpretação dos resultados obtidos.

## **4. Modelos Utilizados**

Nesta etapa foi criado um notebook direcionado para divisão da base de dados em conjuntos de treino e teste como a criação e estruturação da *pipeline* para posterior utilização nos dados de treino. 

<br>

### 4.1. Divisão do *Dataset* em Meses de Verão e Inverno

Após a importação dos dados em formato *parquet* do notebook anterior, foram preparados os dados para a modelação. Em primeiro lugar foi dividido o dataset em dados relativos aos meses de verão e aos meses de inverno, através da variável binária **is_summer**. 

<br>

### 4.2.  Seleção das variáveis para a Modelação

De seguida foi feita uma seleção das variáveis que serão usadas para a modelagem:

- **household_id**;
- **energy_median**;
- **energy_mean**;
- **energy_max**;
- **energy_std**;
- **energy_sum**;
- **energy_min**;
- **tax_type**;
- **status**.

No *dataset*, existe para cada **household_id**, uma contabilização da energia utilizada para cada dia, assim como um tipo de tarifa, grupo *Acorn* e o seu respetivo *status*. Decidiu-se, numa fase seguinte, agrupar os datasets de meses de verão e de inverno por **household_id**, onde a informação sobre a contabilização de energia de cada dia foi agrupada e feita a média das observações e as restantes, sendo apenas selecionadas as categorias de cada um. Foram criadas as variáveis **avg_energy_median**, **avg_energy_mean**, **avg_energy_max**, **avg_energy_std**, **avg_energy_sum** e  **avg_energy_min**.

Cada base de dados ficou com um número semelhante de linhas porque em todas as casas presentes, foi identificada informação em ambos os meses.

<br>

### 4.3.  Divisão das Bases de Dados em Conjuntos de Treino e Teste

De seguida procedeu-se à divisão em conjuntos de treino e teste. Para cada base de dados, foi feita uma divisão de 80% para dados de treino e 20% para dados de teste.

<br>

### 4.4.  Criação e Aplicação da 1ª Pipeline: Codificação das variáveis **tax_type** e  **status**

Para a esta primeira *pipeline*, foram consideradas as seguintes fases:

1. **Indexação (Indexer)**: Criação de novas variáveis que irão converter as categorias das  variáveis categóricas (**tax_type** **status**) para índices numéricos, criando as variáveis  **status_indexed** e **tax_type_indexed**;
2. **Codificação (Encoder)**: Criação de novas variáveis que irão converter os índices previamente criados nas variáveis  **status_indexed** e **tax_type_indexed** e convertê-los em vetores binários, criado as variáveis **Status_encoded** e **tax_type_encoded**;
3. **Atribuição (Assembler)**: Selecionar as variáveis que irão contribuir diretamente para a determinação do clustering, convertendo-as num único vetor de *features*. As variáveis utilizadas são: **avg_energy_median**, **avg_energy_mean**, **avg_energy_max**, **avg_energy_std**, **avg_energy_sum**,  **avg_energy_min**, **Status_encoded** e **tax_type_encoded**. Estas serão convertidas numa única variável chamada **unscaled_features**;
4. **Estandardização (Scaler)**: Utiliza a variável **unscaled_features** previamente obtida e para cada linha, subtrai a média e divide pelo desvio-padrão. É criada uma nova variável a partir dessa chamada **scaled_features**.

Esta pipeline foi  aplicada nos dados de treino de ambas as bases de dados.


<br>

### 4.5.  Criação da 2ª Pipeline:  Aplicação do Modelo K-means

No modelo *K-means*, é necessário imputar manualmente inicialmente o valor de *k*. Como tal foi implementado um avaliador (*ClusteringEvaluator*),  que permitiu avaliar e identificar o melhor número de clusters. Chama-se *Silhouette Score* e calcula um valor que classifica a adequação do *Clustering* realizado, para cada número de *k* dado. 

Para ambos os datasets, o valor de *k* ótimo é 11, o que pode indicar que existem 11 formas diferentes de consumo, de acordo com os critérios definidos previamente. Estas 11 formas diferentes de consumo podem possivelmente indicar os padrões de consumo dos consumidores

Por fim, foi aplicada a 2ª Pipeline (a aplicação do modelo *K-means*) com o *k* ótimo para cada dataset, classificando cada **household_id** num cluster.

<br>

### 4.6.  Determinação do melhor valor de *k* e aplicação do 2ª Pipeline

No modelo K-means, é necessário imputar manualmente inicialmente o valor de *k*. Como tal foi implementado um avaliador (*ClusteringEvaluator*) criterização que permite avaliar e identificar o melhor número de clusters. Chama-se *Silhouette Score* e calcula um valor que classifica a adequação do *Clustering* realizado, para cada número de *k* dado. Para cada base de dados, para valores de k entre 2 e 18,  foi aplicado o modelo K-means e determinado o *k* com maior *Silhouette Score*. 

Para ambos os datasets, o valor de *k* ótimo é 11, o que pode indicar que existem 11 formas diferentes de consumo, de acordo com os critérios definidos previamente. Estas 11 formas diferentes de consumo podem possivelmente indicar os padrões de consumo dos consumidores

Por fim, foi aplicado o 2ª Pipeline (a aplicação do modelo K-means) de cada *k* ótimo para cada dataset, determinando cada cluster para cada **household_id**.

Aqui, como temos 3 categorias definidas, essa divisão de 2 níveis não fica tão evidente em termos visuais, no entanto, percebe-se que, se se dividisse a categoria *‘Comfortable’* na ordem dos 10 kwh, daria para associar a metade do lado esquerdo com os clusters da categoria *‘Affluent’*, e a metade do lado direito com os clusters da categoria *‘Adversity’*. Fazendo essa associação obter-se-ia uma divisão similar à do gráfico anterior.

Mais uma vez, o *‘ACORN-U’* não se encaixa em nenhum destes grupos, sendo uma categoria em si próprio.

### 5.5. Análise de Utilidade
 
Estes resultados podem ser bastante úteis para os governos na implementação de medidas que promovam a sustentabilidade ambiental. Ao estudar as diferentes características das residências dentro do mesmo grupo *Acorn* e entender o que as diferencia em termos de consumo energético é possível nivelar o consumo energético pela média ou pelo valor mais baixo.

Isso pode ser feito ao identificar práticas e tecnologias que são implementáveis para cada grupo, promovendo sua adoção. Dessa forma, os governos podem introduzir medidas que visem a diminuição sustentável do consumo energético, adaptadas às capacidades econômicas conhecidas de cada grupo *Acorn*.


<br>

#### Medidas para Grupos com Maiores Capacidades Financeiras (Acorn A-E):

##### - Incentivos Fiscais para Energias Renováveis

Oferecer deduções fiscais para a instalação de painéis solares, sistemas de aquecimento solar e outras tecnologias de energia renovável.
Visto que estes grupos têm uma maior capacidade financeira para investir em tecnologias de energia renovável. Incentivos fiscais podem acelerar a adoção de painéis solares e sistemas de aquecimento solar, contribuindo para a redução do consumo energético e das emissões de carbono.

##### - Subsídios para Tecnologias de Eficiência Energética

Proporcionar subsídios para a compra de eletrodomésticos de alta eficiência energética, como frigoríficos e máquinas de lavar de baixo consumo. 
Mesmo com uma alta capacidade financeira, subsídios podem incentivar ainda mais a compra de eletrodomésticos eficientes, reduzindo o consumo energético total dessas residências.

##### - Programas de Educação e Consciencialização

Implementar campanhas de consciencialização sobre o uso eficiente de energia e os benefícios das tecnologias limpas, incentivando práticas de consumo sustentável.
Educar estes grupos sobre práticas de consumo sustentável pode aumentar o impacto das tecnologias limpas.

<br>

#### Medidas para Grupos com Capacidades Financeiras Médias (ACORN F-J):

##### - Programas de Financiamento Acessível:

Oferecer programas de financiamento com baixos juros para a instalação de tecnologias de eficiência energética, como janelas de isolamento térmico.
Estes grupos podem não ter capital imediato para grandes investimentos em eficiência energética, por isso, oferecer financiamento acessível permite que melhorem as suas residências sem um grande impacto financeiro inicial.

##### - Incentivos para Reformas Energéticas:

Proporcionar incentivos para reformas domésticas que melhorem a eficiência energética, como a instalação de isolamento térmico e sistemas de aquecimento mais eficientes.
Incentivos específicos para reformas podem ajudar estes grupos a melhorar a sua eficiência de suas casas de maneira acessível, reduzindo assim as contas de energia a longo prazo.

##### - Descontos em Contas de Energia:

Implementar programas de descontos nas contas de energia para residências que adotam práticas de consumo eficiente.
Reduzir as contas de energia através de descontos pode aliviar a carga financeira desses grupos enquanto que também os incentiva a adotar práticas energéticas mais eficientes.

<br>

#### Medidas para Grupos com Menores Capacidades Financeiras (ACORN K-Q):

##### - Apoio Direto para Melhorias Energéticas:

Fornecer apoio financeiro direto para a substituição de equipamentos antigos e ineficientes por modelos mais eficientes.
Esses grupos muitas vezes não têm recursos para investir em tecnologias eficientes. Apoio financeiro direto pode ajudar a substituir equipamentos antigos por modelos mais recentes e eficientes, reduzindo significativamente o consumo de energia.

##### - Programas de Eficiência Energética Comunitária:

Desenvolver programas comunitários que promovam a eficiência energética, como workshops sobre economia de energia e distribuição de kits de eficiência energética (lâmpadas LED, temporizadores, etc.).
Iniciativas comunitárias podem alcançar um grande número de pessoas, promovendo práticas de eficiência energética e fornecendo recursos que podem ter um impacto imediato no consumo de energia.

##### - Subsídios para Aquecimento e Climatização:

Oferecer subsídios para a instalação de sistemas de aquecimento e climatização eficientes.
Subsídios específicos para aquecimento e climatização são essenciais para estes grupos, especialmente em áreas como Londres, onde o consumo de energia para conforto térmico pode ser uma carga significativa.


<br>

#### Medidas Gerais para Todos os Grupos:

##### - Regulamentação de Construções Sustentáveis:

Estabelecer normas e regulamentações para a construção de novas habitações que integrem padrões de eficiência energética e sustentabilidade.
Estabelecer normas para construções sustentáveis garante que as novas habitações sejam à *priori* eficientes energeticamente, reduzindo o consumo de energia a longo prazo para todos os grupos.

##### - Promoção de Transportes Públicos e Alternativos:

Investir em infraestruturas de transporte público eficientes e acessíveis, promover o uso de bicicletas e incentivar as caminhadas.
Melhorar os transportes públicos e promover alternativas sustentáveis reduz a dependência de veículos particulares, diminuindo assim as emissões de carbono, promovendo um estilo de vida mais sustentável.

##### - Incentivos para Empresas Sustentáveis:

Criar incentivos para empresas que adotam práticas sustentáveis e tecnologias limpas, reduzindo seu impacto ambiental e promovendo uma economia verde
Incentivar empresas a adotarem práticas sustentáveis pode ter um efeito cascata, reduzindo o impacto ambiental das operações empresariais e promovendo uma economia verde que beneficia toda a sociedade.

Implementar estas medidas pode ajudar a reduzir o consumo energético de forma sustentável e adaptada às características e capacidades financeiras de cada grupo *Acorn*, promovendo a sustentabilidade ambiental de forma inclusiva e eficaz.  É de realçar que, mais im portante, seria estudar as categorias de diferentes clusters dentro dos mesmos acorns, de forma a nivelar os acorns que consomem mais de acordo com os que consomem menos.
